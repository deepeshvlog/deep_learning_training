```python
#filters
noisy_img=cv2.imread("/content/drive/MyDrive/noisy_image.jpeg",1)
cv2_imshow(noisy_img)
print(noisy_img.shape)

kernel=np.ones((3,3),dtype=np.float32)/9

denoised_img1=cv2.filter2D(noisy_img,-1,kernel)
cv2_imshow(denoised_img1)
```
